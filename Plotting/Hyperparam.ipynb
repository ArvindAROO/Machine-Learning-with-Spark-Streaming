```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../src/performance/hyperparam/supervised-run1.csv')
df
```

```python
plt.plot(df['SGD L1 Accuracy'], color='r', label='L1 loss')
plt.plot(df['SGD L2 Accuracy'], color='b', label='L2 loss')
plt.legend()
plt.title('Loss function comparision for SGD(SVM)')
plt.xlabel('Batch No.')
plt.ylabel('Accuracy')
plt.show()
```

```python
plt.plot(df['NB alpha 1 Accuracy'], color='r', label='alpha=1')
plt.plot(df['NB alpha 2 Accuracy'], color='b', label='alpha=2')
plt.plot(df['NB alpha 4 Accuracy'], color='g', label='alpha=4')
plt.legend()
plt.title('Naive Bayes Smoothing Factor tuning')
plt.xlabel('Batch No.')
plt.ylabel('Accuracy')
plt.show()
```

```python
plt.plot(df['PA C 1 Accuracy'], color='r', label='C=1')
plt.plot(df['PA C 5 Accuracy'], color='b', label='C=5')
plt.legend()
plt.title('Passive Agressive Regularization/Penalty tuning')
plt.xlabel('Batch No.')
plt.ylabel('Accuracy')
plt.show()
```
